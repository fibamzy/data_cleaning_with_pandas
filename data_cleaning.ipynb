# Data Cleaning in Pandas

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Customer Call List.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates()
data
```

```python
# modifying the format of the data in last_name col:
# lstrip - takes off LHS, and vice versa for rstrip

data['Last_Name'] = data['Last_Name'].str.lstrip('...')
data
```

```python
data['Last_Name'] = data['Last_Name'].str.lstrip('/')
data
```

```python
data['Last_Name'] = data['Last_Name'].str.rstrip('_')
data
```

```python
# faster way of handling with strip is:

# data['Last_Name'].str.strip('123._/') # this way you list out the character formats you want taken out from LHS & RHS
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# adjusting the format for phone number section:
# step 1 - unify formatting

data['Phone_Number'] = data['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
data
```

```python
# step 2 - convert datatypes from float to strings

data['Phone_Number'] = data['Phone_Number'].apply(lambda x: str(x))
data['Phone_Number']
```

```python
# step 3 - implement preferred new formatting:

data['Phone_Number'] = data['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
data['Phone_Number']
```

```python
# step 4 - adjust the NaN


data['Phone_Number'] = data['Phone_Number'].str.replace('nan--','')
data['Phone_Number'] = data['Phone_Number'].str.replace('Na--','')

data['Phone_Number']
```

```python
data
```

```python
# splitting address:

data['Address'].str.split(',',2, expand=True)
```

```python
# creating new sub cols for the split addresses:

data[['Street_Address', 'State', 'Zip Code']] = data['Address'].str.split(',',2,expand=True)
data
```

```python
# modifying paying customer and do not contact cols:

data['Paying Customer'] = data['Paying Customer'].str.replace('Yes', 'Y')
data['Paying Customer'] = data['Paying Customer'].str.replace('No', 'N')
data['Do_Not_Contact'] = data['Do_Not_Contact'].str.replace('Yes', 'Y')
data['Do_Not_Contact'] = data['Do_Not_Contact'].str.replace('No', 'N')
data
```

```python
# sorting NaN values across dataset:

data = data.fillna('')
data
```

```python
# sorting out the accounts to be contacted by the call centre:
# 1. remove the ones spcified as do not contact

for x in data.index:
    if data.loc[x, 'Do_Not_Contact']== 'Y':
        data.drop(x, inplace=True)
        
data
```

```python
# 2. remove rows with blank phone numbers:

for x in data.index:
    if data.loc[x, 'Phone_Number']== '':
        data.drop(x, inplace=True)
        
data

# alternative and faster way:
# data = data.dropna(subset='Phone_Number', inplace=True) - works when dropping null values
```

```python
# reseting index:
data = data.reset_index(drop=True) # adding drop, removes the original index
```

```python
data = data.drop(['Address', 'Not_Useful_Column','Unnamed: 8', 'Unnamed: 9'], axis=1)
```
